```python
# dependencies

import dns.resolver
import pandas as pd
```

```python
# Example of querying a domain's MX records

domain = 'accelo.com'
for x in dns.resolver.resolve(domain, 'MX'):
    print(x.to_text())

# Sample responses
# Microsoft 365: accelo.mail.protection.outlook.com
# Google: 
#     aspmx.l.google.com
#     alt1.aspmx.l.gooogle.com
#     aspmx2.googlemail.com
```

```python
# Get the basic user details from a CSV file.  Only required column is titled 'email' and should contain email addresses

df = pd.read_csv('user_email_data.csv')
df.head()
```

```python
# Function for querying a domain's MX records, reviewing each record and categorizing the domain as either using "google", "microsoft365" or "other" to power its email

def domainChecker(domain):
    
    try:
        mxRecords = dns.resolver.resolve(domain, 'MX')
    except Exception as e: 
        print(f'Error: {e}')
        host = 'other'
        return host
    
    for x in mxRecords:
        mxList = x.to_text().split('.')
        for i in mxList:
            if i == 'google':
                host = 'google'
                break
            elif i == 'googlemail':
                host = 'google'
                break
            elif i == 'outlook':
                host = 'microsoft365'
                break
            else:
                host = 'other'
                continue
        
    return host
```

```python
# Loop to process each email address in the CSV and create a list of the addresses' email host
# WARNING: Can require up to 10 seconds per address

rowNum = 0
hostList = []

for i in df['email']:
    print(f'Evaluating row #{rowNum}')
    
    email = str(df['email'][rowNum])
    print(f'Email: {email}')
    if email == 'nan':
        host = ''
    else:
        domain = email.split('@')[1]            
        host = domainChecker(domain)

    hostList.append(host)



    print(f'Host: {host}')
    rowNum+=1
    
print(hostList)
```

```python
# Add the list of hosts to the dataframe as a new column

df['host'] = hostList
```

```python
# Create a CSV from the dataframe

df.to_csv('results.csv', index=False)
```
